```python
import os
import subprocess
from pyngrok import ngrok
import time
import socket

# 1. Nettoyage des processus existants
ngrok.kill()  # Fermer tous les tunnels ngrok actifs
subprocess.run(["pkill", "-f", "streamlit"], stderr=subprocess.DEVNULL)
subprocess.run(["pkill", "-f", "ngrok"], stderr=subprocess.DEVNULL)
time.sleep(2)

# 2. Lancer Streamlit en arrière-plan
streamlit_cmd = [
    "streamlit", "run", "appp.py",
    "--server.port=8502",
    "--server.headless=true",
    "--server.enableCORS=false"
]
subprocess.Popen(streamlit_cmd)
print("⏳ Lancement de Streamlit sur le port 8502...")

# 3. Attendre que le port 8502 soit prêt
def wait_for_port(port, timeout=30):
    start = time.time()
    while time.time() - start < timeout:
        try:
            with socket.create_connection(("localhost", port), timeout=1):
                return True
        except:
            time.sleep(1)
    return False

if not wait_for_port(8502):
    raise RuntimeError("❌ Échec du démarrage de Streamlit")

# 4. Démarrer Ngrok et afficher le lien public
ngrok.set_auth_token("2xGzVfMW8zf8369kTLMtC7FVqVp_4NTXiM7JHWJmu75KoXWnm")
public_url = ngrok.connect(8502, bind_tls=True).public_url
print(f"\n✅ Application Streamlit disponible ici : {public_url}\n")
```
